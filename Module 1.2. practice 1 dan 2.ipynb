    1. Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

```python
import numpy as np
import matplotlib.pyplot as plt

# Inisialisasi dataset
Gol_Terbanyak = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
    
print(f"Jumlah Gol: {Gol_Terbanyak}")

# Mean
mean_Gol = sum(Gol_Terbanyak) / len(Gol_Terbanyak)
print("Mean Gol:", mean_Gol)

# Range
range_Gol = max(Gol_Terbanyak) - min(Gol_Terbanyak)
print("Range Gol:", range_Gol)

# Variance and Standard Deviation
variance_Gol = sum((x - mean_Gol) ** 2 for x in Gol_Terbanyak) / len(Gol_Terbanyak)
std_deviation_Gol = variance_Gol ** 0.5
print("Variance Gol:", variance_Gol)
print("Standard Deviation Gol:", std_deviation_Gol)

plt.hist(Gol_Terbanyak, bins=20, edgecolor='black', alpha=0.5, color='purple')

# Define the horizontal line coordinates
y = 1.5 # The y-coordinate where the line will be drawn
x1 = 4/70*1  # The x-coordinate of the start point
x2 = 66/70*1  # The x-coordinate of the end point

# Add a horizontal line
plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {range_Gol}')

# Add vertical line
plt.axvline(x=min(Gol_Terbanyak), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(Gol_Terbanyak):.2f}')
plt.axvline(x=max(Gol_Terbanyak), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(Gol_Terbanyak):.2f}')
plt.axvline(x=mean_Gol, color='black', linestyle='dashed', linewidth=1, label=f'Mean = {mean_Gol}')
plt.axvline(x=mean_Gol - std_deviation_Gol, color='brown', linestyle='dashed', linewidth=1, label=f'{mean_Gol-std_deviation_Gol:.2f}')
plt.axvline(x=mean_Gol + std_deviation_Gol, color='brown', linestyle='dashed', linewidth=1, label=f'{mean_Gol+std_deviation_Gol:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range')
plt.title(f'Std Deviation = {std_deviation_Gol:.2f}')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
